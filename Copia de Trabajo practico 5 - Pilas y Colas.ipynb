### **Ejercicio 1**

Implementar el TDA Pila (Stack), con las siguientes operaciones:
- Crear (__init__())
- Vaciar
- Apilar elemento (push)
- Desapilar elemento (pop)
- Obtener primer elemento (top)
- Clonar
- Obtener tamaño de pila
- Esta vacía.
- __repr__(). Para poder imprimir una Pila por consola

```
class Pila:
  def __init__(self):
    self.items=[]

  def vaciar(self):
    self.items.clear()

  def apilar(self):
    self.items.append()

  def desapilar(self):
   self.items.pop()

  def primerElemento(self):
     self.items.top()
  
  def clonar(self):
    self.items.copy()
  
  def tamanyo(self):
    return len(self.items)

  def __repr__(self):
    return str(self.items)

  

  

 
    
    
```
